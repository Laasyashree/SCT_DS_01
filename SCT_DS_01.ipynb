```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("world_population.csv")

print("\n--- Dataset Head ---")
print(df.head())

print("\n--- Dataset Info ---")
df.info()

# Set seaborn style
sns.set(style="whitegrid")

# -------- Bar Chart: Number of Countries by Continent --------
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Continent", order=df["Continent"].value_counts().index, palette="viridis")
plt.title("Number of Countries by Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------- Histogram: Distribution of 2022 Population --------
plt.figure(figsize=(10, 6))
sns.histplot(df["2022 Population"], bins=30, kde=True, color='skyblue')
plt.title("Distribution of 2022 Population")
plt.xlabel("2022 Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()

# -------- Histogram: Growth Rate --------
plt.figure(figsize=(10, 6))
sns.histplot(df["Growth Rate"], bins=30, kde=True, color='salmon')
plt.title("Distribution of Growth Rate")
plt.xlabel("Growth Rate")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()

# -------- Histogram: Area (km²) --------
plt.figure(figsize=(10, 6))
sns.histplot(df["Area (km²)"], bins=30, kde=True, color='lightgreen')
plt.title("Distribution of Country Area (km²)")
plt.xlabel("Area (km²)")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
